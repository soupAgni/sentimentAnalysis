```python
import pandas as pd
import numpy as np
```

```python
drumpf_df = pd.read_csv("drumpf_data.csv")
drumpf_df = drumpf_df.loc[:, ~drumpf_df.columns.str.contains('^Unnamed')]
drumpf_df['Handle'] = "realDonaldDrumpf"
print drumpf_df.shape


trump_df = pd.read_csv("data.csv")
trump_df = trump_df.loc[:, ~trump_df.columns.str.contains('^Unnamed')]
trump_df['Handle'] = "realDonaldTrump"
print trump_df.shape

frames = [trump_df, drumpf_df]
all_tweets = pd.concat(frames)
print list(all_tweets)
all_tweets
```

```python
all_tweets['Tweet'] = all_tweets['Tweet'].str.replace('[^\w\s]','')
all_tweets
```

```python
all_tweets.drop(["Date", "Retweets"], axis = 1, inplace = True)
all_tweets.head()
```

```python
all_tweets = all_tweets.sample(frac=1)
all_tweets.tail()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_tweets["Tweet"], all_tweets["Handle"], test_size=0.25, random_state=42, stratify=all_tweets["Handle"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tvec = TfidfVectorizer(stop_words='english',ngram_range=(1, 2), min_df = .025, max_df=.95)
tvec.fit(X_train.values.astype('U'))
tweet_train = pd.DataFrame(tvec.transform(X_train.values.astype('U')).todense(), columns=tvec.get_feature_names())
tweet_test = pd.DataFrame(tvec.transform(X_test.values.astype('U')).todense(), columns=tvec.get_feature_names())
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
# Setup logistic regression and score train set
logreg = LogisticRegression()
logreg.fit(tweet_train, y_train)
print "Train score: ", logreg.score(tweet_train, y_train)
print
#using logreg on test set
y_probs = logreg.predict(tweet_test) #predict y values for X_test
print "Output for Tested Model: "
print "Confusion Matrix of Predictions: "
print
print(confusion_matrix(y_test, y_probs))
print
#printing classification report
print "Classification Matrix: "
print
print(classification_report(y_test, y_probs, target_names=["Drumpf", "Trump"]))
print "Test score: ", logreg.score(tweet_test,y_test)
```
